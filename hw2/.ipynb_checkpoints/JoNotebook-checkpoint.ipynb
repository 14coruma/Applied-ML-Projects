```python
# Divide into k = 10 folds on Batch Gradient Descent
# Reference for K-folds : course slides + https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.KFold.html
# Reference for Batch Gradient Descent : https://ozzieliu.com/2016/02/09/gradient-descent-tutorial/ 

from sklearn.model_selection import KFold
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error # loss function

lr = LinearRegression()
skfolds = KFold(n_splits = 10, shuffle=True, random_state = 42)

eta = 0.0001 # learning rate
niter = 200 # number of iterations

plt.figure(figsize=(20, 2))
plt.title('MSE of Batch Gradient Descent of each K-fold')
plt.ylabel('MSE of K-fold')
plt.xlabel('Iteration')
for train_index, test_index in skfolds.split(X, y):
    theta = np.random.randn(4,1)

    #print("TRAIN:", train_index, "TEST:", test_index)
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    x_new_mat = X_train
    x0 = [1] * len(X_train)
    x_new_mat.insert(loc = 0, column=0, value = x0)

    #print(x_new_mat)
    MSE_loss_iter = [0] * niter

    for iter in range(niter):
        y_hat = x_new_mat.dot(theta)

        MSE_loss_iter[iter] = mean_squared_error(y_train, y_hat) # calculate MSE

        N = len(y_train)

        gradients = 2/N * x_new_mat.T.dot(x_new_mat.dot(theta) - y_train) # calculate gradient
        theta = theta - (eta * gradients) # update theta
    plt.plot(np.arange(niter), MSE_loss_iter)
    plt.yscale("log")
```

```python
# Stochastic Gradient Descent
# Reference : https://neuraspike.com/blog/linear-regression-stochastic-gradient-descent-python/ 

def next_batch(features, labels, batch_size):
    """
    :param features: feature matrix
    :param labels: target vector
    :param batch_size:  size of mini-batch 
    :return: a list of the current batched features and labels
    """
    batch_list = []
    # iterate though a mini batch of both our features and labels
    for data in np.arange(start=0, stop=np.shape(features)[0], step=batch_size):
        # append the current batch
        batch_list.append((features[data: data + batch_size],
                           labels[data: data + batch_size]))
 
    return batch_list
```

```python
# Mini-batch Gradient Descent : test for learning rate and batch size
# Reference for mini-batch gradient descent : https://www.geeksforgeeks.org/ml-mini-batch-gradient-descent-with-python/ 

# function to create a list containing mini-batches 
def create_mini_batches(X, y, batch_size): 
    mini_batches = [] 
    data = np.hstack((X, y)) 
    np.random.shuffle(data) 
    n_minibatches = data.shape[0] // batch_size 
    i = 0
  
    for i in range(n_minibatches + 1): 
        mini_batch = data[i * batch_size:(i + 1)*batch_size, :] 
        X_mini = mini_batch[:, :-1] 
        Y_mini = mini_batch[:, -1].reshape((-1, 1)) 
        mini_batches.append((X_mini, Y_mini)) 
    if data.shape[0] % batch_size != 0: 
        mini_batch = data[i * batch_size:data.shape[0]] 
        X_mini = mini_batch[:, :-1] 
        Y_mini = mini_batch[:, -1].reshape((-1, 1)) 
        mini_batches.append((X_mini, Y_mini)) 
    return mini_batches 
  
# function to perform mini-batch gradient descent 
def miniGradientDescent(X, y, eta, batch_size, niter): 
    theta = np.zeros((X.shape[1], 1)) 
    MSE_loss_MB = [] 

    for iter in range(niter): 
        mini_batches = create_mini_batches(X, y, batch_size) 
        MB_loss = []
        for mini_batch in mini_batches: 
            X_mini, y_mini = mini_batch 
            h = hypothesis(X, theta)
            gradient = np.dot(X.transpose(), (h - y))
            theta = theta - eta * gradient

            loss = np.dot((h - y).transpose(), (h - y))
            loss /= 2
            MB_loss.append(loss[0])
        MSE_loss_MB.append(np.average(MB_loss))
  
    return theta, MSE_loss_MB

[theta_MB, MSE_loss_MB] = miniGradientDescent(X, y, eta = 0.0001, batch_size = 5, niter = 200)
plt.figure(figsize=(20, 2))
plt.title('MSE of Mini Batch Gradient Descent')
plt.ylabel('MSE by MB iteration')
plt.xlabel('Iteration')

plt.plot(np.arange(niter), MSE_loss_MB)
plt.yscale("log")
```
