A hospital staff member wants to determine if a patient's satisfaction with the hospital (y) can be predicted from the patient's age (x1, in years), severity of illness (x2), and anxiety level (x3). Forty-six patients were randomly selected, where the data is in the file patient_satisfaction.txt with column 1 containing y-values, and columns 2-4 containing values for x1, x2, and x3, respectively. Large values indicate more satisfaction, illness severity, and anxiety. For this problem:

```python
# Get data as variables
# Reference for reading in data : https://www.symbiosisacademy.org/tutorial-index/pandas-read_csv-custom-separators/
# Reference for reading in DF columns : https://www.earthdatascience.org/courses/intro-to-earth-data-science/scientific-data-structures-python/pandas-dataframes/indexing-filtering-data-pandas-dataframes/ 

import pandas as pd
df = pd.read_csv('patient_satisfaction.txt', sep='\s+', header  = None) # read data
#print(df)

# get values y, x1, x2, and x3
y = df.iloc[:, 0:1]
print(y)

x1 = df.iloc[:, 1:2]
#print(x1) # age

x2 = df.iloc[:, 2:3]
#print(x2) # severity

x3 = df.iloc[:, 3:4]
#print(x3) # anxiety

X = df.iloc[:, 1:4] 
#print(X) # for all Xs
```

```python
# Reference for title to subplots : https://stackoverflow.com/questions/25239933/how-to-add-title-to-subplots-in-matplotlib

import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()

scat_x1 = fig.add_subplot(131)
scat_x2 = fig.add_subplot(132)
scat_x3 = fig.add_subplot(133)

### create nicer plots

scat_x1.scatter(y, x1)
scat_x1.title.set_text('Satisfaction vs. Age')

scat_x2.scatter(y, x2)
scat_x2.title.set_text('Satisfaction vs. Severity')

scat_x3.scatter(y, x3)
scat_x3.title.set_text('Satisfaction vs. Anxiety')
```

```python
# Divide into k = 10 folds on Batch Gradient Descent
# Reference for K-folds : course slides + https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.KFold.html
# Reference for Batch Gradient Descent : https://ozzieliu.com/2016/02/09/gradient-descent-tutorial/ 

from sklearn.model_selection import KFold
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error # loss function

lr = LinearRegression()
skfolds = KFold(n_splits = 10, shuffle=True, random_state = 42)

eta = 0.0001 # learning rate
niter = 100 # number of iterations

BGD_MSE = []
BGD_theta = []

plt.figure(figsize=(20, 2))
plt.title('MSE of Batch Gradient Descent of each K-fold')
plt.ylabel('MSE of K-fold')
plt.xlabel('Iteration')
for train_index, test_index in skfolds.split(X, y):
    theta = np.random.randn(4,1)

    #print("TRAIN:", train_index, "TEST:", test_index)
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    x_new_mat = X_train
    x0 = [1] * len(X_train)
    x_new_mat.insert(loc = 0, column=0, value = x0)

    #print(x_new_mat)
    MSE_loss_iter = [0] * niter

    for iter in range(niter):
        y_hat = x_new_mat.dot(theta)

        MSE_loss_iter[iter] = mean_squared_error(y_train, y_hat) # calculate MSE

        N = len(y_train)

        gradients = 2/N * x_new_mat.T.dot(x_new_mat.dot(theta) - y_train) # calculate gradient
        theta = theta - (eta * gradients) # update theta
    # predicting output for X_test using the final theta
    x00 = [1] * len(X_test)
    X_test.insert(loc = 0, column = 0, value = x00)
    y_pred = np.dot(X_test, theta)
    error = mean_squared_error(y_test, y_pred)
    BGD_MSE.append(error)
    BGD_theta.append(theta)

    plt.plot(np.arange(niter), MSE_loss_iter)
    plt.yscale("log")

print("MSE = ", np.mean(BGD_MSE)) # print out mean MSE across K-folds for Batch Gradient Descent
print("Mean Theta = ", np.mean(BGD_theta, axis = 0)) # print mean Theta across K-folds for the last iterations
```

Stochastic Gradient Descent

```python
# Stochastic Gradient Descent
# Reference : https://neuraspike.com/blog/linear-regression-stochastic-gradient-descent-python/ 

def next_batch(features, labels, batch_size):
    batch_list = []
    # iterate though a mini batch of both our features and labels
    for data in np.arange(start=0, stop=np.shape(features)[0], step=batch_size):
        # append the current batch
        batch_list.append((features[data: data + batch_size],
                           labels[data: data + batch_size]))
 
    return batch_list
```

```python
# stochastic gradient descent : test for learning rate effect
batch_size = 1
niter_SGD = 100
eta_SGD = 0.0001

SGD_MSE = []
SGD_theta = []

plt.figure(figsize=(20, 2))
plt.title('MSE of Stochastic Gradient Descent of each K-fold')
plt.ylabel('MSE of K-fold')
plt.xlabel('Iteration')

for train_index, test_index in skfolds.split(X, y):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    m = np.shape(X_train)[0]  # total number of samples
    n = np.shape(X_train)[1]  # total number of features

    X_train_new = X_train
    X_train_new = np.concatenate((np.ones((m, 1)), X_train_new), axis=1)
    theta_SGD = np.random.randn(n + 1, 1) # random initial weights
    
    MSE_loss_iter_SGD = []

    for iter in np.arange(niter_SGD):  
        batch_epoch_loss = []
    
        for (X_batch, y_batch) in next_batch(X_train_new, y_train, batch_size):
            batch_m = np.shape(X_batch)[0]
            y_estimated = X_batch.dot(theta_SGD)
            error = y_estimated - y_batch 
            cost = (1 / 2 * batch_m) * np.sum(error ** 2)
            batch_epoch_loss.append(cost) 
            gradient = (1 / batch_m) * X_batch.T.dot(error)
            theta_SGD = theta_SGD - eta_SGD * gradient
    
        MSE_loss_iter_SGD.append(np.average(batch_epoch_loss))

    x00 = [1] * len(X_test)
    X_test.insert(loc = 0, column = 0, value = x00)
    y_pred = np.dot(X_test, theta_SGD)
    error = mean_squared_error(y_test, y_pred)
    SGD_MSE.append(error)
    SGD_theta.append(theta_SGD)

    plt.plot(np.arange(niter_SGD), MSE_loss_iter_SGD)
    plt.yscale("log")

print("MSE = ", np.mean(SGD_MSE)) # print out mean MSE across K-folds for SGD
print("Mean Theta = ", np.mean(SGD_theta, axis = 0)) # print out theta across K-folds for the last iterations
```

In summary, at 100 iterations ot the gradient descent process, the three algorithms - batch gradient descent (BGD), stochastic gradient descent (SGD), and mini-batch gradient descent (MBGD) - were able to calculate thetas for linear regressions using an eta of 0.0001.

So when comparing the MSE of the algorithms at eta = 0.0001 with 100 iterations,
BGD had
MSE =  636.0295391385287

SGD had
MSE =  311.8710035665864

MBGD, with a batch size of 10 had
MSE =  330.02625621707415

MBGD, with a batch size of 20 had
MSE =  393.72149708866834

MBGD, with a batch size of 5 had
MSE =  311.72861220907555

MBGD, with a batch size of 2 had
MSE =  308.7

MBGD, with a batch size of 1 (which is in principal, identical to SGD except for the inherent stochastic "randomness") had
MSE =  330.7

So, while we tested out various learning rates for each algorithm, the best algorithm given a fixed learning rate of 0.0001, would be to use mini-batch gradient descent with a batch size of 2, based on mean squared error. Also based on this algorithm's results, the theta of this dataset's linear regression is bias = 0.45714764, w1 = -1.83788653, w2 = 2.67664329, w3 = -0.30626588.

However, other measures such as "processing speed" might be another useful measure in addition to MSE for researchers when applying the method to much larger datasets. In general, mini-batch gradient descent algorithms tend to be a middle solution in computational burden and performance when compared to batch or stochastic gradient descent algorithms.
