```python
import numpy as np
import pandas as pd
from sklearn.datasets import fetch_openml
from sklearn.cluster import KMeans
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
mnist = fetch_openml('mnist_784', version=1)
```

```python
print(mnist.keys())
#df = pd.DataFrame(data=np.c_[mnist['data'], mnist['target']],
#                  columns= ['data', 'target'])
print(type(mnist))
```

```python
X, y = mnist["data"], mnist["target"]
y = y.astype(int)
```

```python
# normalizing/standardizing
X = ((X/255.) - .5)*2
print(X.shape, y.shape)
```

```python
#kmeans = KMeans(n_clusters=10, init='random', random_state=10).fit(sample)
#print(kmeans.cluster_centers_)

# How our centroids should ideally look
fix, ax = plt.subplots(nrows=2, ncols=5, sharex=True, sharey=True)
ax = ax.flatten()
for i in range(10):
  matchlist = [i for i, x in enumerate(y==i) if x]
  img = X[matchlist[0]].reshape(28,28)
  ax[i].imshow(img,cmap='Greys')

ax[0].set_xticks([])
ax[0].set_yticks([])
plt.show()
```

```python
# randomly sample 10k out of 70k
indexes = np.random.random_sample(12000,)*70000 # made room for testing, didn't need it
indexes = indexes.astype(int)
sample = X[indexes[:10000]]
labels = y[indexes[:10000]]
```

```python
from sklearn.decomposition import PCA
#pca = PCA(n_components = 2)
#sample = pca.fit_transform(sample) # was giving me issues with displaying results, removed it
```

```python
# Problem 1
distortions = []
for i in range(5,16):
  kmeans = KMeans(n_clusters=i, init='random', n_init=10, max_iter=300, random_state=10)
  kmeans.fit(sample)
  distortions.append(kmeans.inertia_)

plt.plot(range(5,16), distortions, marker='o')
plt.xlabel('Num of clusters')
plt.ylabel('Distortion')
plt.tight_layout()
plt.show()
```

```python
'''
So the elbow method didn't reveal much because even if we have more or less than 10 clusters, which
is logically what we should be choosing as there's 10 digits and 10 categories, a few digits
look pretty similar so they're attributes are near each other and the one or two more or less clusters
won't have a big difference on the total euclidian distances to the clusters but it will make a big
difference when it comes to our model's accuracy. We'll roll with 10 clusters.
'''
```

```python
# Problem 2
%matplotlib inline
kmeans = KMeans(n_clusters=10, init='random', random_state=10).fit(sample) # random is default
centroids = kmeans.cluster_centers_
#print(centroids.shape)
#print(type(centroids))
centroids = ((centroids/2.) + 5.) * 255 # return to grayscale so we can plot
#print(centroids.shape)
fig, ax = plt.subplots(nrows=2, ncols=5, sharex=True, sharey=True)
ax = ax.flatten()
for i, centroid in enumerate(centroids):
  img = centroid.reshape(28,28)
  ax[i].imshow(img, cmap='Greys')
```

```python
'''
These clusters perform okay, the numbers aren't in the right spots but that's not a huge deal if
i'm not mistaken. The 0, 6, 2, 3, and 8 labels are pretty clear. We have a couple labels that both
look like 1 and two that look like 9, we have a label that almost looks like a combination of a 3 and an 8, and we have a label
that looks like something between a 8 and a 9. This makes sense because these numbers can look
similarly when written by hand. We're also missing labels for 4 and 5.
'''
```

```python
# Problem 3
distortions = []
for i in range(5,16):
  kmeans = KMeans(n_clusters=i, init='k-means++', n_init=10, max_iter=300, random_state=100)
  kmeans.fit(sample)
  distortions.append(kmeans.inertia_)

plt.plot(range(5,16), distortions, marker='o')
plt.xlabel('Num of clusters')
plt.ylabel('Distortion')
plt.tight_layout()
plt.show()
```

```python
kmeansplus = KMeans(n_clusters=10, init='k-means++', random_state=10).fit(sample)
centroids = kmeansplus.cluster_centers_
#print(centroids.shape)
centroids = ((centroids/2.) + 5.) * 255 # return to grayscale so we can plot
#print(centroids.shape)
fig, ax = plt.subplots(nrows=2, ncols=5, sharex=True, sharey=True)
ax = ax.flatten()
for i, centroid in enumerate(centroids):
  img = centroid.reshape(28,28)
  ax[i].imshow(img, cmap='Greys')
```

```python
'''
Here we have pretty clear labels for 2, 0, 6, 3, and 8.
There's two labels for 1 and two for 9, and we have another label that looks like a mix of a
8 and a 9.
We're also missing labels for 4, 5, and 7, so I can't say this clustering algorithm is performing
well. There's a few solid centroids, but largely doesn't perform great.
'''
```
